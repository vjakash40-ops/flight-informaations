```python
from flask import Flask, render_template_string

app = Flask(__name__)
```

# Flight Data

```python

flights = [
    {"id": 1, "name": "IndiGo 6E101", "from": "Delhi", "to": "Mumbai", "price": 5000, "seats": 5},
    {"id": 2, "name": "Air India AI202", "from": "Chennai", "to": "Bangalore", "price": 3000, "seats": 3},
    {"id": 3, "name": "SpiceJet SG303", "from": "Hyderabad", "to": "Kolkata", "price": 4500, "seats": 4}
]
```

# HTML Template

```python

html = """
<!DOCTYPE html>
<html>
<head>
    <title>Flight Details</title>
    <style>
        body { font-family: Arial; background-color: #f4f4f4; text-align: center; }
        table { margin: auto; border-collapse: collapse; width: 80%; background: white; }
        th, td { padding: 10px; border: 1px solid #ddd; }
        th { background-color: #007BFF; color: white; }
        h1 { color: #333; }
    </style>
</head>
<body>
    <h1>Available Flights</h1>
    <table>
        <tr>
            <th>ID</th>
            <th>Flight Name</th>
            <th>From</th>
            <th>To</th>
            <th>Price (â‚¹)</th>
            <th>Seats</th>
        </tr>
        {% for f in flights %}
        <tr>
            <td>{{ f.id }}</td>
            <td>{{ f.name }}</td>
            <td>{{ f.from }}</td>
            <td>{{ f.to }}</td>
            <td>{{ f.price }}</td>
            <td>{{ f.seats }}</td>
        </tr>
        {% endfor %}
    </table>
</body>
</html>
"""

@app.route("/")
def home():
    return render_template_string(html, flights=flights)

app.run()
```
